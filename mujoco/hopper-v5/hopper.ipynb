```python
import gymnasium as gym
```

```python
env = gym.make("Hopper-v5", render_mode="human")
```

```python
state_dim = env.observation_space.shape[0]
action_dim = env.action_space.shape[0]

state_dim, action_dim
```

action space는 3개로 이루어져있음. 첫번째는 허벅지 로터에 적용되는 토크 (허벅지 관절) / 두번째는 다리 로터에 적용되는 토크 (다리 관절) / 세번째는 발 로터에 적용되는 토크 (발 관절)

observation space는 11개로 이루어져있음. 
5개는 파트들의 position value이고, 6개는 각 파트들의 속도이다.
